```python
# Import required modules
# pandas for data manipulation and analysis
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("dataset.csv")
```

```python
# Display the first 5 rows of the dataset
df.head()
```

```python
# Display dataset information
df.info()
```

## Preproccessing Steps

```python
# Fill missing values with '0'
df = df.fillna(0)

# Display the first 5 rows of the dataset
df.head()
```

```python
# Set products count data type to 'in64'
for column in df.iloc[:, 1:]:
    df[column] = df[column].astype("int64")

# Display dataset information
df.info()
```

```python
# Parse the 'Date' column to datetime format
df["date"] = pd.to_datetime(df["date"], format="%m/%d/%Y")

# Display the first 5 rows of the dataset
df.head()
```

```python
# Select 'date' and 'sales_total' columns
time_series_df = df[['date', 'sales_total']].dropna()
```

```python
# Rename columns from 'date' to 'ds' and 'sales_total' to 'y'
time_series_df.rename(columns={"date": "ds", "sales_total": "y"}, inplace=True)
```

## Forecasting with Prophet

```python
# Import prophet
# Prophet is a forecasting tool to predict future values of 
# a time series
from prophet import Prophet
```

```python
# Create a new Prophet model
model = Prophet()
```

```python
# Fit the model to the dataset
model.fit(time_series_df)
```

```python
# Create a new dataframe with 14 days in the future
future = model.make_future_dataframe(periods=31)
```

```python
# Make future predictions
forecast = model.predict(future)
```

```python
# Shows the last few predictions
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

## Plotting Datas

```python
# Import matplotlib for visualization 
import matplotlib.pyplot as plt
```

### Original Sales Total Datas

```python
# Set the plot size
plt.figure(figsize=(20, 10))

# Set the title of the plot
plt.title("Original Sales Total")

# Plot the original data
plt.plot(
    time_series_df["ds"], time_series_df["y"], label="Pre-prediction Data", color="blue"
)

# Set x-axis label to 'Date'
plt.xlabel("Date")

# Set y-axis label to 'Sales Total'
plt.ylabel("Sales Total")

# Show the plot grid
plt.grid()

# Show the plot results
plt.show()
```

### Prediction Sales Total Datas

```python
# Plot the predicted data
model.plot(forecast)

# Set the title of the plot
plt.title("Predicted Sales Total")

# Set x-axis label to 'Date'
plt.xlabel("Date")

# Show the plot legend
plt.legend()

# Set y-axis label to 'Sales Total'
plt.ylabel("Sales Total")
```
